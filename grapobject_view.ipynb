```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=px.data.gapminder()
df
```

```python
fig=px.pie(df,names="continent",values="gdpPercap")
fig.show()
```

```python
fig=px.pie(df,names="continent",values="gdpPercap")
fig.update_traces(textposition="outside")
fig.show()
```

```python
fig=go.Figure(go.Pie(labels=df["continent"],values=df["pop"],hole=0.3))
fig.show()
```

```python
df=px.data.gapminder().query("country=='India'")
df
```

```python
fig=go.Figure(go.Scatterpolar(r=[1,2,4,5,6,7,9],
              theta=[45,70,90,180,250,190,20],
              mode="text",text=['a','b','c',4,5,6,7]))
fig.show()
```

```python
from plotly.subplots import make_subplots
```

```python
fig=make_subplots(rows=1,cols=3,subplot_titles=('P1','P2','P3'))
fig.append_trace(go.Scatter(x=df['year'],y=df['pop'],
                            mode='text+markers+lines',
                            text=df['gdpPercap']),row=1,col=1)
fig.append_trace(go.Bar(x=df['year'],y=df['pop']),row=1,col=2)
fig.append_trace(go.Scatter(x=df['country'],y=df['pop']),row=1,col=3)
fig.update_layout(title_text='Subplots')
fig.show()
```

```python
fig=make_subplots(rows=2,cols=2,subplot_titles=('P1','P2','P3','P4'),
                  specs=[[{'type':'xy'},{'type':'polar'}],[{'type':'scene'},{'type':'domain'}]])
fig.append_trace(go.Scatter(x=df['year'],y=df['pop'],mode='text+markers+lines',text=df['gdpPercap']),row=1,col=1)
fig.append_trace(go.Barpolar(r=[1,2,4,5,6,7,9],
                             theta=[45,70,90,180,250,190,20],width=20),row=1,col=2)
fig.append_trace(go.Scatter3d(x=[1,2,3,4,5],y=[10,20,30,40,50],z=[100,200,300,400,500]),row=2,col=1)
fig.append_trace(go.Pie(values=[45,89,90]),row=2,col=2)
fig.update_layout(title_text='Subplots')
fig.show()
```
